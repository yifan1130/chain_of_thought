# Visualize model performance against model scale

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
model_names = ["Minerva 540B", "PaLM 540B", "PaLM 64B", 
               "LLaMA 65B", "LLaMA 33B", "LLaMA 13B", "FlanT5 11B", "LLaMA 7B", "FlanT5 3B"]
model_scale = [540, 540, 64, 65, 33, 13, 11, 7, 3]
model_perf = [58.8, 56.9, 52.4, 50.9, 35.6, 17.8, 16.1, 11, 13.5]

fig = plt.figure(figsize=(5, 4), dpi=200)
for i in range(len(model_names)):
    plt.scatter(model_scale[i], model_perf[i], 
                label=model_names[i], color="C%d" %i,
                s=200
                )
plt.xlabel("Model Scale")
plt.ylabel("GSM8k Accuracy")
plt.xscale("log")

# Fit linear regression
slope, intercept, r_value, p_value, std_err = linregress(np.log10(model_scale), model_perf)
x = np.linspace(1, 600, 100)
y = slope * np.log10(x) + intercept

# Plot regression line
plt.plot(x, y, color='black', linestyle='--')

plt.axhline(y=92, linestyle='--', color='#1f77b4')
plt.text(3.5, 92.5, "GPT-4", color='#1f77b4')

plt.axhline(y=81.8, linestyle='--', color='#1f77b4')
plt.text(3.5, 81.8, "Claude-v1.3", color='#1f77b4')

plt.axhline(y=80.7, linestyle='--', color='#1f77b4')
plt.text(3.5, 80.7-5, "PaLM-2", color='#1f77b4')

plt.axhline(y=74.9, linestyle='--', color='#1f77b4')
plt.text(3.5, 74.9-5, "GPT-3.5-Turbo", color='#1f77b4')

plt.axhline(y=66.6, linestyle='--', color='#1f77b4')
plt.text(3.5, 67-5, "Code-davinci-002", color='#1f77b4')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
