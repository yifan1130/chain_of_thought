```python
def find_answer(s):
    assert('boxed' in s)
    ans = s.split('boxed')[-1]
    if(ans[0] == '{'):
        stack = 1
        a = ''
        for c in ans[1:]:
            if(c == '{'): 
                stack += 1
                a += c
            elif(c == '}'): 
                stack -= 1
                if(stack == 0): break
                a += c
            else: 
                a += c
    else:
        a = ans.split('$')[0].strip()
    return a
```

```python
find_answer(math_algebra_by_level['Level 1'][2]['solution'])
```

```python
def sample(math_algebra_by_level_k):
    cases = np.random.choice(len(math_algebra_by_level_k), 3, replace=False)
    for ci in cases:
        print(math_algebra_by_level_k[ci]['solution'])
        print()
        print(find_answer(math_algebra_by_level_k[ci]['solution']))
        print()
        print('--')
```
